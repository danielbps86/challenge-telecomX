#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'

df = pd.read_json(url)

df
```

```python
import requests
import json

response = requests.get(url)

data=''
if response.status_code == 200:
    data = json.loads(response.text)
    #print(data)
else:
    print(f"Error retrieving data, status code: {response.status_code}")

df=pd.json_normalize(data)
df
```

#ðŸ”§ TransformaciÃ³n

```python
#VerificaciÃ³n de datos nulos
df.isnull().sum()
```

```python
#VerificaciÃ³n de tipo de dato de cada columna
df.info()
```

```python
#VerificaciÃ³n de datos NaN
df.isna().sum()
```

```python
#VerificaciÃ³n de valores Ãºnicos
for key in df.keys():
    print(key,':' ,df[key].nunique())
    if df[key].nunique()<30:
      print(pd.unique(df[key]))
      print('-'*60)
```

```python
#Verificar duplicados
print(df.duplicated().sum())
```

```python
#Verificar donde hay valores vacÃ­os
df.apply(lambda x: x.astype(str).str.strip() =='').sum()
```

```python
#CreaciÃ³n de DF con datos vacÃ­o en Churn para visualizaciÃ³n y posible correciÃ³n de errores
df_empty = df[df['Churn'] == '']
df_empty
```

```python
#Reemplazar valores vacÃ­os en Churn por N/A
df['Churn'] = df['Churn'].replace('', 'N/A')
```

```python
#CreaciÃ³n de DF con datos vacÃ­os en account.Charges.Total para visualizaciÃ³n y posible correciÃ³n de errores
df_empty2 = df[df['account.Charges.Total'] == ' ']
df_empty2
```

```python
#Reemplazar valores vacÃ­os en account.Charges.Total por 0
df['account.Charges.Total'] = df['account.Charges.Total'].replace(' ', 0)

#Verificar valores vacÃ­os nuevamente
df.apply(lambda x: x.astype(str).str.strip() =='').sum()
```

```python
#VerificaciÃ³n de valores Ãºnicos nuevamente
for key in df.keys():
    print(key,':' ,df[key].nunique())
    if df[key].nunique()<30:
      print(pd.unique(df[key]))
      print('-'*60)
```

```python
#ModificaciÃ³n de los tipos de datos de la columna account.Charges.Total
df['account.Charges.Total'] = df['account.Charges.Total'].astype(float)
```

```python
#Agregar columna 'Cuentas_Diarias'de tipo float64 al df
df['Cuentas_Diarias'] = df['account.Charges.Monthly'] / 30

df.info()
```

#ðŸ“Š Carga y anÃ¡lisis

```python
#DescripciÃ³n de los datos
df.describe()
```

```python
#DistribuciÃ³n de la deserciÃ³n de clientes
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

valores = list(df['Churn'].value_counts())
totalValores=sum(valores)
desercion = {'Desercion':valores}
indices = ['No','SÃ­','N/D']

ddff= pd.DataFrame.from_dict(desercion)
ddff.index = indices
ddff.head()
#barplot
paleta = sns.color_palette("hls", 3)
sns.barplot(x=ddff.index, y=ddff['Desercion'],palette=paleta)
plt.title('DeserciÃ³n de clientes')
#axis labels
plt.xlabel('DeserciÃ³n')
plt.ylabel('Cantidad')
#bars text
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width()/2., p.get_height(),
             '{:1.0f}%'.format(100*p.get_height()/totalValores),
             ha="center")
plt.show()
```

```python
#Recuento de evasiÃ³n por variables categÃ³ricas
#Data frame con Churn = 'Yes' y customer.gender
total_clientes_desertados = len(df[df['Churn']=='Yes'])
df_gender=df[df['Churn']=='Yes']
df_gender=df_gender['customer.gender'].value_counts()
df_gender.index = ['Mujer','Hombre']

totalVal = df_gender.values.sum()#Suma de valores con Churn =='Yes'

#Data frame con Churn = 'Yes' y account.Contract
df_contract = df[df['Churn']=='Yes']
df_contract=df_contract['account.Contract'].value_counts()
df_contract.index = ['Mes a mes','Un aÃ±o','Dos aÃ±os']
df_contract.columns = ['Tipo de Contrato','Cantidad']
#df_contract


#Data frame con Chrun='Yes' y account.PaymentMethod
df_paymethod = df[df['Churn']=='Yes']
df_paymethod=df_paymethod['account.PaymentMethod'].value_counts()
df_paymethod.index = ['Cheque electrÃ³nico','Cheque por correo','Transferencia bancaria','Tarjeta de crÃ©dito']
#print(df_paymethod)

fig, axes = plt.subplots(1, 3, figsize=(24,6))
fig.suptitle('DeserciÃ³n de clientes por categorÃ­as')
plt.subplots_adjust(wspace=0.5, hspace=0.5)
sns.barplot(y=df_gender.index, x=df_gender.values,ax=axes[0],orient='h', width=0.5,palette=sns.color_palette("rocket",2), hue=df_gender.index,legend=False)
axes[0].set_title('DeserciÃ³n de clientes por gÃ©nero')
axes[0].set_ylabel('GÃ©nero')
axes[0].set_xlim(0,2000)
sns.barplot(y=df_contract.index, x=df_contract.values,ax=axes[1],orient='h',width=0.5,palette=sns.color_palette("mako",3),hue=df_contract.index,legend=False)
axes[1].set_xlim(0,2000)
axes[1].set_title('DeserciÃ³n de clientes por tipo de contrato')
axes[1].set_ylabel('Tipo de contrato')
sns.barplot(y=df_paymethod.index, x=df_paymethod.values,ax=axes[2],orient='h',width=0.5,palette=sns.color_palette("flare",4),hue=df_paymethod.index,legend=False)
axes[2].set_xlim(0,2000)
axes[2].set_title('DeserciÃ³n de clientes por mÃ©todo de pago')
axes[2].set_ylabel('MÃ©todo de pago')
#bars text
for i in range(len(axes)):
  for p in axes[i].patches:
    axes[i].text(p.get_width() + 110, p.get_y() + p.get_height()/2,
                 '{:1.1f}%'.format(100*p.get_width()/totalVal),
             ha="center")




plt.show()
```

```python
#Conteo de evasiÃ³n por variables numÃ©ricas
fig, axes = plt.subplots(1, 2, figsize=(24,6))
plt.subplots_adjust(wspace=0.5, hspace=0.5)
sns.histplot(data=df, x="account.Charges.Total", hue="Churn", multiple="stack",binwidth=1000,palette=sns.color_palette("mako",3),ax=axes[0])
axes[0].set_title('DeserciÃ³n de clientes por monto de facturaciÃ³n')
axes[0].set_ylabel('Cantidad de clientes')
axes[0].set_xlabel('Monto total de facturaciÃ³n ($)')
axes[0].set_ylim(0,3000)
axes[0].set_xlim(0,9000)

sns.histplot(data=df, x="customer.tenure", hue="Churn", multiple="stack",binwidth=12,palette=sns.color_palette("rocket",3),ax=axes[1])
#modificar eje X a 6
axes[1].set_xlim(0,78)
axes[1].set_ylim(0,3000)
#modificar escala eje X
axes[1].set_xticks(np.arange(0, 78, 12))
axes[1].set_title('DeserciÃ³n de clientes por tiempo como cliente')
axes[1].set_ylabel('Cantidad de clientes')
axes[1].set_xlabel('Meses como cliente')
plt.show()
```
